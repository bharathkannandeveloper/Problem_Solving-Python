```python
# Creating Series Using :
"""
1) List
2) Tuple
3) Set
4) Dictionary
5) Using General Method ( Direct Method )
"""

import pandas as pd

list1=[1,2,3,4,5]
series=pd.Series(list1)
print("\nConverting List to Series : \n",series)
tuple1=(6,7,8,9,10)
series=pd.Series(tuple1)
print("\nConverting to Tuple to Series : \n",series)
set1=('Rahul','Varma','Arjun','Das')
series=pd.Series(set1)
print("\nConverting Set to Series : \n",series)
dict1={'Rahul':20,'Varma':30,'Arjun':21,'Das':26}
series=pd.Series(dict1)
print("\nConverting Dictionary to Series : \n",series)
series = pd.Series([1,2,3,4,5])
print("\n Using Direct Pandas Series : \n",series)
```

```python
# Indexing In the Pandas - series

import pandas as pd

list1=[1,2,3,4,5]
series=pd.Series(list1,index=['a','b','c','d','e'])
print("Original Series\n")
print(series)

#Using Lable Indexing (loc())
series=series.loc['a':'c']
print("\nAfter Indexing\n")
print(series)

list1=[1,2,3,4,5]
series=pd.Series(list1)
print("Original Series\n")
print(series)

#Using Actual Indexing (loc())
series=series.iloc[0:2]
print("\nAfter Indexing\n")
print(series)
```

```python
# Slicing With Series

import pandas as pd

series = pd.Series([1,2,3,4,5,6])
#Using the Normal [] Method
a=series[1:3]
print(a)
# Using iloc() function
a=series.iloc[1:3]
print(a)
```

```python
# Arithmetic Operations with Series 
#( Addition , Subtraction , Multiplication , Division , Exponential , Modulo )

import pandas as pd 
list1=[1,2,3,4,5]
list2=[6,7,8,9,10]
series1=pd.Series(list1)
series2=pd.Series(list2)
print("-------------Direct Method --------------\n")
# Direct Method
# Addition
add_series=series1+series2
print("Addition\n")
print(add_series)
# Subtraction
sub_series=series1-series2
print("\nSubtraction\n")
print(sub_series)
#Multiplication
mul_series=series1*series2
print("\n Multiplication \n")
print(mul_series)
#Division
print("\n Division \n")
div_series=series1/series2
print(div_series)
#Exponential
expo_series=series1**series2
print("\n Exponential \n")
print(expo_series)
#Modulo
mod_series=series1%series2
print("\n Modulo \n")
print(mod_series)


print("\n ---------------- using Functions -------------\n")
#using Functions 
# Addition
add_series=series1.add(series2)
print("Addition\n")
print(add_series)
sub_series=series1.sub(series2)
print("\nSubtraction\n")
print(sub_series)
mul_series=series1.mul(series2)
print("\n Multiplication \n")
print(mul_series)
div_series=series1.div(series2)
print("\n Division \n")
print(div_series)
expo_series=series1.pow(series2)
print("\n Exponential \n")
print(expo_series)
mod_series=series1.mod(series2)
print("\n Modulo \n")
print(mod_series)

```

```python
# Comparison Operation In Pandas Series to Filter the Data By Certian Conditions
# Getting Boolean Output

import pandas as pd

list1=[1,4,2,4,35,56,43,5,2]
list2=[3,2,3,46,35,23,4,5,2]
series1=pd.Series(list1)
series2=pd.Series(list2)

# Using The Comparison Operations

# Greater than
greaterthen=series1>series2
print("\n Greaterthen \n")
print(greaterthen)

# Lesser then 
lesserthen=series1<series2
print("\n Lesserthen \n")
print(lesserthen)

# Greater than or Equal to
great_eqal=series1>=series2
print("\n Greaterthen or Equal \n")
print(great_eqal)

#Lesser then or Equal to
lesser_eqal=series1<=series2
print("\n Lesserthen or Equal \n")
print(lesser_eqal)

#Equavalent
equalent = series1==series2
print("\n Equavalent \n")
print(equalent)
```

```python
# Aggregation In Series - it Means to Combining Multiple Values into The single Value By Applying Functions
# Ex : max() , Min() , count() , mean() , median() , 

import pandas as pd

list1=[1,2,3,4,5,6]
series1=pd.Series(list1)

# Applying Max
#maxi=max(series1) - Also Use this
maxi=series1.max()
print("\n Maximum \n")
print(maxi)

# Apply min
#mini = min(series1) - Also use this
mini=series1.min()
print("\n Minimum \n")
print(mini)

# Apply count
#count = len(series1) - Also use this
count=series1.count()
print("\n Length or Count \n")
print(count)

# Apply mean
mean= series1.mean()
print("\n Mean \n")
print(mean)

# Apply Median
median = series1.median()
print("\n Median \n")
print(median)

# Apply Sum
sum1=series1.sum()
print("\n Sum of Elements \n")
print(sum1)
```

```python
# Pandas Series Methods
  
import pandas as pd
list1=[1,2,3,4,5,6,None,7,8,9,10,10]
series1=pd.Series(list1)
series2=pd.Series(['Rahul','Kowshik','Bharath','Aditya'])

# Using head() Method to Print first n Number of Row or Columns (Note: Default 5)
print("\n printing series with Default Head\n")
print(series1.head())

# Using tail() Method to Print last n Number of Row or Columns (Note: Default 5)
print("\n printing series with Default Head\n")
print(series1.tail())

#Using Describe Method
print("\n printing series with Descibe Method describe() \n")
print(series1.describe())

#Using Info method to get Information
print("\n Getting Information of Series using Info() Method - Get Datatype and Memory \n ")
print(series1.info())

#Using Mean Method
print("\n Mean of Series")
print(series1.mean())

#Using Unique Method
print("\n Use unique() Method to get only unique Numbers and Avoid Duplicates \n")
print(series1.unique())

# Using Value_counts() Method . its used to count a each number with how many Duplicates and print each Count
print("\n Use value_counts() Method \n")
print(series1.value_counts())

# using the sort_values() to sort the Series
print("\n Use sort_values() for Sorting to print in ascending order \n")
print(series1.sort_values())
print("\n Use sort_values(ascending=False) for Sorting to print in Descending order \n")
print(series1.sort_values(ascending=False))

# Using Apply() Method to Apply function to each element of the series
print("\n use apply() Method - To apply function to each elements in Array")
def add(x):
    x=x+100
    return x
print(series1.apply(add))

# Using fillna() Method to Fill the Values where its Null or Empty
print("\n use fillna() Method to Fill Missing Values")
print(series1.fillna("Empty"))

# Using drop() Method to Remove the Specific row or Column from the Series
print("\n Using Drop() Method to Remove row or Columns")
print(series1.drop(6))

# Using concat() Method to Concatenate two Series
print("\n using Concatenate Method to Concatenate two series\n")
print(pd.concat([series1,series2]))
# Ignoring The Indexing
print("\n using Concatenate Method to Concatenate two series and Ignore the Indexing \n")
print(pd.concat([series1,series2],ignore_index=True))
```

```python
# Creating Data Frames Using Different Ways 

import pandas as pd

#Creating Empty Data Frame
df=pd.DataFrame()
print("\n Empty Data frame \n")
print(df)

# Creating Dataframe using a List
list1=[['Rahul',30],['Kowshik',21],['Randy',34]]
df=pd.DataFrame(list1)
print("\n Creating Dataframe using a List \n")
print(df)

#Creating a Dataframe using a Dictionary
dict1={'Name':['Bharath','Rahul','Varun','Varma','Rohit'],'Age':[20,32,19,24,30]}
print("\n Creating a Dataframe using a Dictionary \n")
df=pd.DataFrame(dict1)
print(df)

# Create a Df using the List and Dictionary Inside
list1=[
       {'Name':'Bharath','Age':20},
       {'Name' :'Varma','Age':28},
       {'Name' :'Rohit','Age':30},
      ]
print("\n Create a Df using the List and Dictionary Inside \n")
df=pd.DataFrame(list1)
print(df)
```

```python
# Select , Add and Delete Columns

import pandas as pd
list1={'Name':['Bharath','Kamal','Rohit','Varsha','Priya'],'Age':[20,30,24,28,19]}
df=pd.DataFrame(list1)
print("\n Original Data frame \n")
print(df)

# Selecting the Columns
column1=df['Name']
print("\n Selecting The Column \n")
print(column1)

# Adding the Columns
df['Mobile']=[1234567890,8987654321,63761387,983918398,9232779823]
print("\n Adding The new Column in the Existing Data Frame \n ")
print(df)

# Deleting the Column in the Data Frame
df=df.drop('Mobile',axis=1)
print("\n Deleting the Column from The Data Frame \n")
print(df)
```

```python
# Select , Add and Delete Row From the Data Frame

import pandas as pd
list1={'Name':['Bharath','Kamal','Rohit','Varsha','Priya'],'Age':[20,30,24,28,19]}
df=pd.DataFrame(list1)
print("\n Original Data frame \n")
print(df)

# Selecting a row
row = df.iloc[0]
print("\n Selecting the Row \n")
print(row)

# Adding the Row in the Data Frame
row = {'Name':'Gokul','Age':22}
df.loc[len(df)] = row
print("\n Adding the new Row \n")
print(df)

# Deleting the Row in the data Frame
df=df.drop(len(df)-1) 
print("\n Deleting the Row in the Data Frame  \n")
print(df)
```

```python
# Filtering the Data frame

import pandas as pd
data={'Name':['Bharathkannan','Rahul','Varun','Pooja','Veera'],
      'Age' : [20,30,29,19,21],
      'Marks':[90,20,35,30,42],
     }
df=pd.DataFrame(data)
print("\n Original data Frame \n")
print(df)

# Filtering the Pass mark in df
print("\n Filtering the pass Marks From Data \n")
marks = df[df['Marks']>= 35]
print(marks)

# Filtering the fail Marks in df
print("\n Filtering the fail Marks From Data \n")
marks = df[df['Marks'] < 35]
print(marks)

```

```python
# Using Functions in Pandas - [ Series - map() and Data frames - Applymap() ]

import pandas as pd

# Using applymap() in Data Frame

data = {'Age' : [21,22,28,19,23],
        'Percentage' : [99,80,75,97,86],
       }
df=pd.DataFrame(data)
print("\n Original Data Frame \n")
print(df)

def func(x):
    return x**2

print("\n After Applying function \n")
print(df.applymap(func))

# using the map()  Function in the Series

data2 = [1,2,3,4,5,6,7,8,9]
series=pd.Series(data2)
print("\n Original Series \n")
print(series)

def func(x):
    return x**2

print("\n After Applying function \n")
print(series.map(func))
```

```python
import pandas as pd

data = {'Age': [20, 30, 29, 19, 21],
        'Marks': [100, 99, 90, 89, 92],
        }

df = pd.DataFrame(data)
print("\n Original data \n")
print(df)

# Row or Column -wise Operation
def row(x):
    return x.sum()

# Apply Column Wise Operation
print("\n After Apply Cloumn Wise operation \n")
print(df.apply(row,axis=0))

# Applying row-wise operation to each row
print("\n After Apply Row Wise operation \n")
print(df.apply(row, axis=1))
```
