```python
# 1- Linkedlist - Add Elements - At beginning , At End and Inbetween

class Node:
    def __init__(self, data):
        self.data = data
        self.ref = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def addfirst(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.ref = self.head
            self.head = new_node
            
    def addlast(self,data):
        new_node=Node(data)
        if (self.head==None):
            New_node=self.head
        else:
            n=self.head
            while n.ref is not None:
                n=n.ref
            n.ref=new_node
    
    def display(self):
        if self.head is None:
            print("The Linked List is Empty")
        else:
            n = self.head
            while n is not None:
                print(n.data, end=" --------->  ")
                print(n.ref)
                n = n.ref

    def addaft(self, data, x):
        n = self.head
        new_node = Node(data)
        while n is not None:
            if x == n.data:
                break
            n = n.ref
        if n is None:
            print("The element is Not found in the Linked List")
        else:
            new_node.ref = n.ref
            n.ref = new_node
            
    def addbef(self, data, x):
        new_node = Node(data)
        
        if self.head is None:
            print("The Linked List is Empty")
            return
        
        if self.head.data == x:
            new_node.ref = self.head
            self.head = new_node
            return
        
        n = self.head
        while n.ref is not None:
            if n.ref.data == x:
                break
            n = n.ref
            
        if n.ref is None:
            print("The Element is Not in the Linked List")
        else:
            new_node.ref = n.ref
            n.ref = new_node

LL1 = LinkedList()
LL1.addfirst(500)
LL1.addfirst(400)
LL1.addfirst(200)
LL1.addfirst(100)
LL1.addaft(300, 200)
LL1.addbef(600, 100)
LL1.addlast(800)
LL1.display()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.ref = None

class LinkedList:
    def __init__(self):
        self.head = None
        
    def addfirst(self, data):
        new_node = Node(data)
        new_node.ref = self.head
        self.head = new_node
        
    def addlast(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            n = self.head
            while n.ref is not None:
                n = n.ref
            n.ref = new_node
    
    def remfirst(self):
        if self.head is None:
            print("The Linked List is Empty, Can't Remove")
        else:
            self.head = self.head.ref
            
    def remlast(self):
        if self.head is None:
            print("The Linked List is Empty, Can't Remove")
        else:
            if self.head.ref is None:
                self.head = None
            else:
                n = self.head
                while n.ref.ref is not None:
                    n = n.ref
                n.ref = None
                
    def rembet(self, x):
        if self.head is None:
            print("The Linked List is Empty, Can't Remove")
        else:
            n = self.head
            while n.ref is not None:
                if n.ref.data == x:
                    n.ref = n.ref.ref
                    return
                n = n.ref
            print(f"Element {x} not found in the Linked List")
            
    def display(self):
        if self.head is None:
            print("The Linked List is Empty")
        else:
            n = self.head
            while n is not None:
                print(n.data, end=" --> ")
                print(n.ref)
                n = n.ref
            print("None")

LL1 = LinkedList()
LL1.addfirst(100)
LL1.addfirst(200)
LL1.addfirst(300)
LL1.addfirst(400)
LL1.remfirst()
LL1.remlast()
LL1.addlast(400)
LL1.addlast(500)
LL1.addlast(600)
LL1.rembet(500)
LL1.display()
```

# Doubly Linked List ...

<span style="font-size: 20px;color: red;">Doubly Linked List - Adding Elements at Begining , Ending and Inbetween </span>

```python
#-----------------------------------------------------------  Creating Node Class  ---------------------------------------------------------------------#

class Node:
    
    def __init__(self, data):
        self.data = data
        self.pref = None
        self.nref = None
        
#-----------------------------------------------------------  Doubly Linkedlist Class  -----------------------------------------------------------------#

class DoublyLinkedlist:
    
    def __init__(self):       
        self.head = None
        self.tail = None
        
    def display(self):
        if self.head is None:
            print("The Linked List is Empty ")
        else:
            n = self.head
            while n is not None:
                print(n.pref.data if n.pref else "None", end=" <--- ")
                print(n.data, end=" ---> ")
                print(n.nref.data if n.nref else "None")
                n = n.nref
            print("-------------------------------\n\n")
                
    def revdisplay(self):
        if self.head is None:
            print("The Linked List is Empty")
        else:
            n = self.head
            while n.nref is not None:
                n = n.nref
            while n is not None:
                print(n.nref.data if n.nref else "None", end=" <--- ")
                print(n.data, end=" ---> ")
                print(n.pref.data if n.pref else "None")
                n = n.pref
            print("-------------------------------\n\n")
            
    def addfirst(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.nref = self.head
            self.head.pref = new_node
            self.head = new_node
            
    def addlast(self, data):
        new_node = Node(data)
        if self.head is None:
            print("The Linked List is Empty")
        else:
            n = self.head
            while n.nref is not None:
                n = n.nref
            new_node.pref = n
            n.nref = new_node
            new_node.nref = self.tail
            
    def insertempty(self, data):
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
            
        else:
            print("The Linked List is Not Empty")
            
    def insertbefore(self,data,x):
        if (self.head==None):
            print("The Linked List is Empty")
        else:
            n=self.head
            while n is not None:
                if(x==n.data):
                    break
                n=n.nref    
            if(n is None):
                print("The Element Not Found")
            else:
                new_node=Node(data)
                new_node.nref=n
                new_node.pref=n.pref
                n.nref=new_node
                    
    def insertafter(self,data,x):
        if (self.head==None):
            print("The Linked List is Empty")
        else:
            n=self.head
            while n is not None:
                if(x==n.data):
                    break
                n=n.nref
            if(n is None):
                print("The Element Not Found")
            else:
                new_node=Node(data)
                new_node.pref=n
                new_node.nref=n.nref
                n.nref=new_node
                
        
LL1 = DoublyLinkedlist()
LL1.insertempty(100)
LL1.addlast(200)
LL1.addlast(300)
LL1.addlast(500)
LL1.insertbefore(400,500)
LL1.addlast(700)
LL1.insertafter(600,500)
LL1.display()
LL1.revdisplay()
```
